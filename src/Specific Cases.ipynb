```julia
using PGFPlots
import HDF5, JLD
using EncounterSimulation
using EncounterModel
using EncounterVisualization
import PyPlot
```

```julia
trl = JLD.load("../data/trlcons_lagrange_sweep_Mar-18_1637.jld")
turning = JLD.load("../data/turning_lagrange_sweep_Mar-18_0608.jld")
trlstatic = JLD.load("../data/D_bisection_Mar-18_1315.jld")
opt_policy=trl["policies"][1]
static_policy=ConstPolicy(HeadingHRL(200))
nothing
```

```julia
data = JLD.load("../data/box_10k_mixed.ic")
ics = data["ics"]
seeds = data["seeds"]
nothing
```

```julia
n = 10000
best=Array(Any,n)
tos = Array(Any,n)
tss = Array(Any,n)
for i in 1:length(best)
    to = EncounterTest(EncounterTestInputData(ics[i],seed=seeds[i],policy=opt_policy))
    ts = EncounterTest(EncounterTestInputData(ics[i],seed=seeds[i],policy=static_policy))
    run!(to,store_hist=false)
    run!(ts,store_hist=false)
    tos[i] = to
    tss[i] = ts
    print("\r$i")
end
```

```julia
for i in 1:length(best)
    to = tos[i]
    ts = tss[i]
    if to.output.deviated && !ts.output.deviated
        best[i] = :static
    elseif ts.output.deviated && !to.output.deviated
        best[i] = :opt
    elseif ts.output.deviated && to.output.deviated
        best[i] = :dev
    else
        best[i] = :none
    end
end
```

```julia
colormap = [:opt=>"g",:static=>"r",:dev=>"k"]
PyPlot.clf()
ax = PyPlot.gca()
for i in 1:length(best)
    if best[i] != :none && best[i] != :dev
        vehicle_arrow(ax, ics[i].is, color=colormap[best[i]], scale=60)
    end
end
```

```julia
is = [1000,1000,3*pi/2]
s = EncounterState([0,0,0],is,false,false)
seeds = 1:10
# static_policy=ConstPolicy(HeadingHRL(trlstatic["Ds"][end]))
static_policy=ConstPolicy(HeadingHRL(200))
opt_policy=trl["policies"][1]
static_ts = [EncounterTest(EncounterTestInputData(s,seed=seed,policy=static_policy)) for seed in seeds]
opt_ts = [EncounterTest(EncounterTestInputData(s,seed=seed,policy=opt_policy)) for seed in seeds]
nothing
```

```julia
run!(static_ts, parallel=false)
run!(opt_ts,parallel=false)
nothing
```

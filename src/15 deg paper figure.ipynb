```julia
using PGFPlots
import HDF5, JLD
using EncounterSimulation
using EncounterModel
```

```julia
function jldhack(fname;range=nothing)
    td = JLD.load(fname,"deviations")
    trr = JLD.load(fname,"risk_ratios")
    if range==nothing
        return ["deviations"=>td,"risk_ratios"=>trr]
    else
        return ["deviations"=>td[range],"risk_ratios"=>trr[range]]
    end
end
trl = jldhack("../data/re_eval_15_Mar-24_1348.jld",range=1:5)
# turning = JLD.load("../data/re_eval_15.0_Mar-24_1321.jld")
# td = JLD.load("../data/re_eval_15.0_Mar-24_1321.jld","deviations")
# trr = JLD.load("../data/re_eval_15.0_Mar-24_1321.jld","risk_ratios")
# turning = ["deviations"=>td,"risk_ratios"=>trr]
turning = jldhack("../data/re_eval_15.0_Mar-24_1321.jld")
trlstatic = JLD.load("../data/D_sweep_15_Mar-24_1304.jld")
curve_data = [
              "Direct control"=>turning,
              "Static TRL"=>trlstatic,
              # "Bounded Static TRL"=>trlstatic2,
              "Optimized TRL"=>trl,
              # "Bounded Optimized TRL"=>trlbounded,
              # "Optimized TRL with D=0 action"=>trlzero,
]
nothing
```

```julia
plots=Plots.Plot[]
# x_param = "avg_delays"
x_param = "deviations"
for (title,d) in curve_data
    p = Plots.Linear(d[x_param],d["risk_ratios"],legendentry=title)
    push!(plots,p)
    @show title
    @show get(d,"lambdas",nothing)
    @show get(d,"Ds",nothing)
    @show get(d,"avg_delays",nothing)
    @show d["deviations"]
    @show d["risk_ratios"]
end
```

```julia
p = Axis(plots,
         xlabel="Deviations",
         ylabel="Risk Ratio",
         legendPos="north east",
         width="3in",
         height="3in")
```

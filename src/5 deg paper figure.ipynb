```julia
using PGFPlots
import HDF5, JLD
using EncounterSimulation
using EncounterModel
```

```julia
trl = JLD.load("../data/trl5_sweep_5_Mar-24_1416.jld")
trl10 = JLD.load("../data/trl5_sweep_5_Mar-24_1542.jld")
trl["deviations"] = [trl10["deviations"][1], trl["deviations"]]
trl["risk_ratios"] = [trl10["risk_ratios"][1], trl["risk_ratios"]]
td = JLD.load("../data/re_eval_5.0_Mar-24_0131.jld","deviations")
trr = JLD.load("../data/re_eval_5.0_Mar-24_0131.jld","risk_ratios")
turning = ["deviations"=>td,"risk_ratios"=>trr]
trlstatic = JLD.load("../data/D_sweep_5_Mar-24_1259.jld")
curve_data = [
              "Direct control"=>turning,
              "Static TRL"=>trlstatic,
              "Optimized TRL"=>trl,
]
nothing
```

```julia
plots=Plots.Plot[]
# x_param = "avg_delays"
x_param = "deviations"
for (title,d) in curve_data
    p = Plots.Linear(d[x_param],d["risk_ratios"],legendentry=title)
    push!(plots,p)
    @show title
    @show get(d,"lambdas",nothing)
    @show get(d,"Ds",nothing)
    @show get(d,"avg_delays",nothing)
    @show d["deviations"]
    @show d["risk_ratios"]
end
```

```julia
p = Axis(plots,
         xlabel="Deviations",
         ylabel="Risk Ratio",
         legendPos="north east",
         width="3in",
         height="3in")
```
